```python
def retrieve_row(df, x, y, z):
    """
    Retrieve the row where column 0 equals x, column 1 equals y, and column 2 equals z.
    
    :param df: DataFrame representing the data.
    :param x: Value to match in column 0.
    :param y: Value to match in column 1.
    :param z: Value to match in column 2.
    :return: The matching row or None if no match is found.
    """
    result = df[(df.iloc[:, 0] == x) & (df.iloc[:, 1] == y) & (df.iloc[:, 2] == z)]
    return result if not result.empty else None

res = retrieve_row(bioDiv_gt, '89e72a749d764c1aacd9284e01c412a4', 5, 11)
```
